```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as scistats
```

```python
data = pd.read_csv("data/data_edited.csv", delimiter=';')
data
```

```python
total_responses = len(data.Q1)-1
print("Number of participants:", total_responses)
```

```python
results_1 = data["Q42"]
results_2 = data["Q43"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')


participants = np.arange(47)
results = {
    'Composition 1': results_1,
    'Composition 2': results_2,
}

x = np.arange(len(participants))  # the label locations
width = 0.5  # the width of the bars
multiplier = 0
offset = 1
fig, ax = plt.subplots(figsize=(16,6),dpi=200, layout='constrained')

for attribute, measurement in results.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, edgecolor="white", label=attribute)
    ax.bar_label(rects, padding=1, fontsize = 14)
    multiplier += 1

ax.set_ylabel('Calificaciones', fontsize = 22)
ax.set_xlabel('Participantes', fontsize = 22)
#ax.set_title('Audience feedback', fontsize = 20)
ax.set_xticks(x + 2* width -0.75, participants, fontsize = 14)
ax.legend(loc='lower left', ncols=1, fontsize = 14, framealpha=1)
ax.set_ylim(0, 11)

plt.savefig('Audience_feedback-THESIS.png')
plt.show()
```

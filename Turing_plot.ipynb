```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data/turing_stats.csv", delimiter=';')
data
```

```python
means = data['Mean']
means = np.array(means)
means
```

```python
data['Std'][0]
```

```python
from matplotlib.ticker import AutoMinorLocator
from matplotlib.patches import Rectangle
```

```python
labels = ['Astronomers','Musicians','Experts','Non experts', 'BVI', 'Non BVI']

fig, ax = plt.subplots(figsize=(14,6),dpi=200)
x = np.arange(len(labels))

bar_width = .9

rects1 = ax.bar(1 - bar_width/2, means[0] , bar_width, yerr=data['Std'][0], 
                label='Astronomers',align='center',ecolor='black',capsize=10)
rects2 = ax.bar(2 - bar_width/2, means[1] , bar_width, yerr=data['Std'][1], 
                label='Musicians',align='center',ecolor='black',capsize=10)
rects3 = ax.bar(3 - bar_width/2, means[2] , bar_width, yerr=data['Std'][2], 
                label='Experts',align='center',ecolor='black',capsize=10)
rects4 = ax.bar(4 - bar_width/2, means[3] , bar_width, yerr=data['Std'][3], 
                label='Non experts',align='center',ecolor='black',capsize=10)
rects5 = ax.bar(5 - bar_width/2, means[4] , bar_width, yerr=data['Std'][4], 
                label='BVI',align='center',ecolor='black',capsize=10)
rects6 = ax.bar(6 - bar_width/2, means[5] , bar_width, yerr=data['Std'][5], 
                label='Non BVI',align='center',ecolor='black',capsize=10)

#ax.set_xticks(x+.55)
#ax.set_xticklabels(labels,size=12)
#ax.set_ylabel('Average success rate',size=25)
ax.yaxis.set_minor_locator(AutoMinorLocator())
#ax.set_ylim([0,100])

#ax.plot([1+bar_width/4, 1+3*bar_width/4],[data['conf_lo'][0],data['conf_hi'][0]],"k--",linewidth=1)


ax.add_patch(Rectangle((.46, data['conf_lo'][0]), 
                       0.18, data['conf_hi'][0]-data['conf_lo'][0],
                       alpha=0.5,color='black',label="Jeffreys' confidence intervals"))
ax.add_patch(Rectangle((1.46, data['conf_lo'][1]), 
                       0.18, data['conf_hi'][1]-data['conf_lo'][1],
                       alpha=0.5,color='black'))
ax.add_patch(Rectangle((2.46, data['conf_lo'][2]), 
                       0.18, data['conf_hi'][2]-data['conf_lo'][2],
                       alpha=0.5,color='black'))
ax.add_patch(Rectangle((3.46, data['conf_lo'][3]), 
                       0.18, data['conf_hi'][3]-data['conf_lo'][3],
                       alpha=0.5,color='black'))
ax.add_patch(Rectangle((4.46, data['conf_lo'][4]), 
                       0.18, data['conf_hi'][4]-data['conf_lo'][4],
                       alpha=0.5,color='black'))
ax.add_patch(Rectangle((5.46, data['conf_lo'][5]), 
                       0.18, data['conf_hi'][5]-data['conf_lo'][5],
                       alpha=0.5,color='black'))
ax.set_xticks([])
ax.set_xticklabels([], size=10)
plt.yticks(fontsize = 20)


#ax.legend(fontsize=10,loc='upper right')
plt.savefig('ACM23_Turing.png')
plt.show()

```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/data_plotting-Balanced.csv", delimiter=';')
data
```

```python
total_responses = len(data)
print("Number of participants:", total_responses)
```

## Qualitative answers

```python
#plot = sns.catplot(data=data, x="Question", y="Qualifications", hue="Group", kind="box")
#with sns.color_palette("bright"):
plot = sns.catplot(
    data=data, x="Questions", y="Qualifications", col="Groups", 
    kind="box", height=8, aspect=.6, col_order=["Experts","Astronomers", "Musicians", "Non experts"],
)
#sns.color_palette("Spectral", as_cmap=True)
sns.set_palette('Spectral')
plot.set(xlabel ="", ylabel = "")
plot.set_titles("{col_name}")

#plot2 = sns.swarmplot(data=data, x="Question", y="Qualifications", size=3)


plt.savefig("Qualitative_analysis.png")
```

```python
qualifications = data["Qualifications"]
qualifications = np.array(qualifications)
```

```python
average = np.mean(qualifications)
median = np.median(qualifications)
print("Average qualification:", average)
print("Median of the qualifications:", median)
```

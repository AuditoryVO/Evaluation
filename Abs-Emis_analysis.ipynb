## 1-Experts vs Non-Experts

## Experts

```python
experts_list = []
experts_index = []
experts_counter = 0
for j in range(len(data)):
    if (((data.Q31[j] == "Musician") and (data.Q21[j] != "Astronomer")) or ((data.Q31[j] == "Musician") and (data.Q21[j] == "Astronomer")) or((data.Q31[j] != "Musician") and (data.Q21[j] == "Astronomer")) and (data.Email[j] != 'RESULTS')):
        experts_counter += 1
        current_row = data.iloc[j]
        experts_list.append(current_row)
        experts_index.append(j)
        
print("Number of experts:", experts_counter)
experts_df = pd.DataFrame(experts_list)
experts_df
```

```python
experts_index
```

## Non experts

```python
nonexperts_list = []
#nonexperts_index = []
nonexperts_counter = 0
for j in range(len(data)):
    if ((data.Q31[j] != "Musician") and (data.Q21[j] != "Astronomer") and (data.Email[j] != 'RESULTS')):
        nonexperts_counter += 1
        current_row = data.iloc[j]
        nonexperts_list.append(current_row)
     #   nonexperts_index.append(j)
        
print("Number of nonexperts:", nonexperts_counter)
nonexperts_df = pd.DataFrame(nonexperts_list)
nonexperts_df
```
